```
# Looking across the rows. Need to drop any?

def look_for_nans_rows(data):
    max_na_row = data.isna().sum(axis=1).max()
    perc_na_row = round(100*max_na_row/data.shape[1],1)
    print(f"\nMax percentage of nans in a row: {perc_na_row}% ({max_na_row})")

    total_na = data.isna().sum().sum()
    perc_na = round(100* total_na/(data.shape[0]*data.shape[1]),1)
    print(f"\nTotal percentage of missing data points: {perc_na}% ({total_na})")

    data.isna().sum(axis=1).hist()
    plt.title('Distribution of missing data across rows')
    plt.show()

look_for_nans_rows(data)
# This plot shows me that nearly 6% of the data is missing, and that most rows
# with missing data are missing only 4-5 data points.
# It also tells me that there are no rows that need to be thrown out, since
# the maximum percentage of data missing (and in only a small minority of rows)
# is less than 20%.
```

```
# Looking down the columns. Should I drop any of these features? Any hierarchical?

def look_for_nans_columns(data):

    print("\nColumns with missing data points:")
    for col in data:
        nmissing = data[col].isna().sum()
        if nmissing > 0:
            print(f"{col} (dtype = {data[col].dtype}): {nmissing} ({round(100*nmissing/data.shape[0])}%) ")
    if data.isna().sum().sum() == 0:
        print('No missing column data.')  
    # print(f"Largest percentage of missing data in column: {round(100*data.isna().sum(axis=0).max()/data.shape[0],2)}")
    # plt.title("Distribution of missing data across columns")
    # plt.show()
    max_na_col = data.isna().sum(axis=0).max()
    perc_na_col = round(100*max_na_col/data.shape[0],1)
    print(f"\nMax percentage of nans in a column: {perc_na_col}% ({max_na_col})")

look_for_nans_columns(data)    
```

```
# Just a function for dilplaying rows with missing values that share a name element
# e.g. "basement" or "garage"

def nans_breakdown(names):

    sim_cols = data.columns[data.columns.str.contains(names)].to_list()
    for col in sim_cols:
        print(f"\n-----{col}-----")
        print(f"dtype: {data[col].dtype}")
        print(f"{col}: {data[col].isna().sum()} NaN's")
        if data[col].dtype == 'object':
            print(f"Unique values in {col}: {data[col].unique()}")
        else:
            print(f'Min = {data[col].min()}, Max = {data[col].max()}')
        if data[col].isna().sum() > 0:
            print(f"A sample of rows where {col} is NaN:\n")
            display(highlight_columns(data.loc[data[col].isna()],columns_to_shadow=col))

def highlight_columns(df, rows=5, color='lightgreen', columns_to_shadow=[], columns_to_show=[]):
    highlight = lambda slice_of_df: 'background-color: %s' % color
    sample_df = df.head(rows)
    highlighted_df = sample_df.style.applymap(highlight, subset=pd.IndexSlice[:, columns_to_shadow])
    return highlighted_df
```

```

def showCategoricalData(data):

    if sum(data.dtypes=='object')>0:

        for col in data[data.columns[data.dtypes=='object']]:
            print(f"\n{col}: {len(data[col].unique())} unique values")
            print(f"     {data[col].unique()}")
    else:
        print("There are no categorical variables in the dataframe.")
showCategoricalData(data)
```

```
# 2 different encodings I'll use for categorical data

def OneHotEncodeColumn(data, data_enc, col_name, drop=True):

    col_enc = pd.get_dummies(data[col_name], drop_first=drop, prefix=col_name)
    if col_name in data_enc.columns:
        data_enc.drop(col_name, axis=1, inplace=True)
    for col in col_enc:
        if col in data_enc.columns:
            data_enc[col] = col_enc[col]
        else:
            data_enc = pd.concat([data_enc, col_enc[col]], axis=1)
    
    display(data_enc[data_enc.columns[data_enc.columns.str.contains(col_name)]].head())

    return data_enc

def LabelEncodeColumn(data, data_enc, col_name, dmap=None):
    from sklearn.preprocessing import LabelEncoder

    if map:
        # use the specified map
        data_enc[col_name] = data[col_name].map(dmap)
    else:
        le = LabelEncoder()
        data_enc[col_name] = le.fit_transform(data[col_name])

    display(data_enc[data_enc.columns[data_enc.columns.str.contains(col_name)]].head())

    return data_enc


def OneHotEncode_CombineColumns(data, data_enc, col_names, new_prefix):

    col0 = OneHotEncodeColumn(data, pd.DataFrame(), col_names[0], drop=False)
    col1 = OneHotEncodeColumn(data, pd.DataFrame(), col_names[1], drop=False)
    conds = data[col_names[0]].append(data[col_names[1]]).unique()

    # Now combine the two in a new df
    comb_df = pd.DataFrame()
    for col in conds:
        if col_names[0]+'_'+col in col0 and col_names[1]+'_'+col in col1:
            comb_df[new_prefix+"_"+col] = np.logical_or(col0[col_names[0]+'_'+col], col1[col_names[1]+'_'+col]).astype(int)
        elif col_names[0]+'_'+col in col0:
            comb_df[new_prefix+"_"+col] = col0[col_names[0]+'_'+col].astype(int)
        else:
            comb_df[new_prefix+"_"+col] = col1[col_names[1]+'_'+col].astype(int)

    # Now add these features back into data_enc (looping to overwrite)
    for col in comb_df:
        data_enc[col] = comb_df[col]

    # Finally, drop the original 2 columns
    for col in col_names:
        if col in data_enc:
            data_enc.drop(columns=col_names, inplace=True)

    return data_enc

quality_dmap = {
    'none':0,
    'Po':1,
    'Fa':2,
    'TA':3,
    'Gd':4,
    'Ex':5,
    }
```

```
def plotPredictions(regr, X_train, X_val, y_train, y_val):

    # Now make predictions!
    y_train_pred = regr.predict(X=X_train)
    y_val_pred = regr.predict(X=X_val)
    # y_test_pred = regr.predict(X=X_test)
    print(f"Train R-squared = {round(regr.score(X=X_train,y=y_train),2)}")
    print(f"Validation R-squared = {round(regr.score(X=X_val,y=y_val),2)}")
    # print(f"Test R-squared = {round(regr.score(X=X_test,y=y_test),2)}")

    # Scatterplot
    plt.figure(figsize=(10,10))
    plt.scatter(y_train, y_train_pred, c='g')
    plt.scatter(y_val, y_val_pred, facecolors='none', edgecolors='k')
    # plt.scatter(y_test,y_test_pred,c='r')

    # plt.legend(['Train','Validation','Test'])
    plt.legend(['Train','Validation'])
    
    # plt.plot([0, max(y_train_pred.max(),y_val_pred.max(),y_test_pred.max())],
    #          [0, max(y_train_pred.max(),y_val_pred.max(),y_test_pred.max())],
    #          c='k',linestyle='--')
    plt.plot([0, max(y_train_pred.max(),y_val_pred.max())],
            [0, max(y_train_pred.max(),y_val_pred.max())],
            c='k',linestyle='--')

    plt.xlabel("Actual Salesprices")
    plt.ylabel("Predicted Salesprices")

    plt.show()
```

```
# load and fit the model
from sklearn.ensemble import RandomForestRegressor

# Load the model
regr = RandomForestRegressor(random_state=0)

# Fit the model
regr.fit(X=X_train,y=y_train)
# print(f"Train R-squared = {round(regr.score(X=X_train,y=y_train),2)}")

plotPredictions(regr, X_train, X_val, y_train, y_val)
```

```
# Fitting an Adaboost model
from sklearn.ensemble import AdaBoostRegressor

regr = AdaBoostRegressor(random_state=0, 
                         n_estimators=100,
                         learning_rate=.5,
                         loss='square')
regr.fit(X_train, y_train)

plotPredictions(regr, X_train, X_val, y_train, y_val)

# Well that wasn't super great. I think these trees are all too shallow? 
# The predictions are oddly trunkated at the bottom, and look strangely binned.
```

```
# Adaboost regression with deeper trees
from sklearn.ensemble import AdaBoostRegressor

mod = RandomForestRegressor(random_state=0,
                            max_depth=25)
regr = AdaBoostRegressor(random_state=0, 
                         n_estimators=200,
                         loss='square',
                         base_estimator=mod)
regr.fit(X_train, y_train)
plotPredictions(regr, X_train, X_val, y_train, y_val)

# mild improvement!
```

```
## Gradient Boosting
from sklearn.ensemble import GradientBoostingRegressor

regr = GradientBoostingRegressor(n_estimators=500,
                                max_depth=5,
                                min_samples_split=2,
                                learning_rate=0.01, 
                                )
regr.fit(X_train,y_train)
plotPredictions(regr, X_train, X_val, y_train, y_val)

```
